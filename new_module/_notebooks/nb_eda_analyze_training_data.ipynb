```python
import pandas as pd
import seaborn as sns
import os
os.chdir('/data/hyeryung/mucoco/')
```

```python
## sentiment 
train_data_yelp = pd.read_json('data/yelp/train.jsonl', lines=True)
```

```python
train_data_yelp['labels'].value_counts().sort_index()
```

```python
train_data_yelp_summary = train_data_yelp['labels'].value_counts().sort_index().reset_index()
```

```python
train_data_yelp_summary['labels'] = train_data_yelp_summary['labels'].astype(str)
```

```python
with sns.color_palette("icefire"):
    sns.histplot(train_data_yelp, x = 'labels')
```

```python
train_data_yelp_summary=train_data_yelp_summary.rename(columns={'labels': 'Ground Truth Positivity', 'count': 'Count'})
with sns.color_palette("vlag"):
    sns.barplot(train_data_yelp_summary, x='Ground Truth Positivity', y='Count', hue='Ground Truth Positivity')
```

```python
train_data_yelp = train_data_yelp.rename(columns={'labels': 'Ground Truth Positivity'})
with sns.color_palette("icefire"):
    sns.histplot(train_data_yelp, x = 'Ground Truth Positivity')
```
