```python
data = []
with open('/home/hyeryung_son/mucoco/outputs/toxicity/locate-edit-gpt2-loc-6toks--1steps-project-1steps/outputs_epsilon-3.txt', 'r') as file:
    for line in file:
        data.append(json.loads(line))

df = pd.json_normalize(data)
```

```python
df['generations.text'] = df['generations'].apply(lambda x: x[0]['text'])
```

```python
df['steps'] = [[] for i in range(len(df))]
```
